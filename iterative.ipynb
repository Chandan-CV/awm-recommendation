```python
import spacy
import pandas as pd
nlp = spacy.load("en_core_web_lg")
```

```python
dataset = pd.read_csv("dataset.csv")
textAnswers = (dataset.iloc[:,3:5]).join(dataset.iloc[:, 8:9]).join(dataset.iloc[:, -2])
dataset.head(10)
```

```python
# print(doc1.similarity(doc2))
# print(doc2.similarity(doc1))
data = textAnswers
# def row_to_text(row):
#     temp =  " ".join(str(row[col]) for col in row.index)
#     doc = nlp(temp)
#     finalString=""
#     for word in doc:
#         if word.pos_ in ['NOUN','VERB','ADJ','ADV'] and word.text != 'nan':
#             finalString += word.text + " "
#     return finalString


def row_to_text(row):
    return " ".join(str(row[col]) for col in row.index)

# Convert each row to a single text string
dataset['combined_text'] = data.apply(row_to_text, axis=1)
```

```python
dataset.head(10)
```

```python

selectedUser = dataset.iloc[9]
selectedUser.values
```

```python
selectedUserText = selectedUser['combined_text']

dataset['similarity'] = dataset['combined_text'].apply(lambda x: nlp(x).similarity(nlp(selectedUserText)))
```

```python
dataset.sort_values(by='similarity', ascending=False, inplace=True)
```
